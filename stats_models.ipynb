```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# load data
data = sm.datasets.get_rdataset("Guerry", "HistData").data
data.head()
```

```python
data.columns
```

```python
# Fit OLS regression use ln
res = smf.ols("Lottery ~ Literacy + np.log(Pop1831)", data=data).fit()
print(res.summary())
```

```python
# Using numpy arrays
# Definir el número de observaciones
nobs = 100

# Generar datos aleatorios X con dimensiones (nobs, 2) utilizando numpy.random.random
X = np.random.random((nobs, 2))

# Agregar una columna de unos a los datos X utilizando sm.add_constant para ajustar un intercepto en el modelo
X = sm.add_constant(X)

# Definir los coeficientes beta del modelo como [1, .1, 0.5]
beta = [1, .1, 0.5]

# Generar errores aleatorios de longitud nobs utilizando numpy.random.random
e = np.random.random(nobs)

# Generar datos de respuesta y utilizando el modelo lineal y = X * beta + e
y = np.dot(X, beta) + e

# Ajustar un modelo de mínimos cuadrados ordinarios (OLS) utilizando los datos generados
# y y X utilizando sm.OLS().fit()
res_2 = sm.OLS(y, X).fit()

# Imprimir un resumen del ajuste del modelo
print(res_2.summary())
```
